```python
# Dependencies and Setup
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#throw in a "hurricane" line for fun even though there weren't any hurricanes in 2015.
```

```python
# Load Hurricane Data
hurricanes = "resources/Hurricane_data_all.csv"

# Load Index data
sp1500 = "resources/S&P1500_Sector_All.csv"
```

```python
# Read the Index and Hurricane csv
sp1500 = pd.read_csv(sp1500)
hurricanes = pd.read_csv(hurricanes)
```

```python
#split the data to only focus on just S&P stocks from 2015

sp1500['Date'] = pd.to_datetime(sp1500['Date'])

sp1500_2015 = sp1500[(sp1500['Date'] > '1/1/2015') & (sp1500['Date'] <= '12/31/2015')]
sp1500_2015 = sp1500_2015.reset_index(drop = True)
```

```python
#rename columns so they'll be easier to call out later

sp1500_2015.columns = ['Date', 'Consumer_Discretionary', 'Communication_Services', 'Consumer_Staples', 
                      'Energy', 'Financials', 'Health_Care', 'Industrials', 'Information_Technology',
                      'Materials', 'Real_Estate', 'Utilities']
sp1500_2015.head()
```

```python
hurricanes
```

```python
# make the line graph with date as the x axis, total return (industry name) as the y axis, add colors afterward

plt.plot(sp1500_2015['Date'], sp1500_2015['Consumer_Discretionary'], label = 'Consumer Discretionary')
plt.plot(sp1500_2015['Date'], sp1500_2015['Communication_Services'], label = 'Communication Services')
plt.plot(sp1500_2015['Date'], sp1500_2015['Consumer_Staples'], label = 'Consumer')
plt.plot(sp1500_2015['Date'], sp1500_2015['Energy'], label = 'Energy')
plt.plot(sp1500_2015['Date'], sp1500_2015['Financials'], label = 'Financials')
plt.plot(sp1500_2015['Date'], sp1500_2015['Health_Care'], label = 'Health Care')
plt.plot(sp1500_2015['Date'], sp1500_2015['Industrials'], label = 'Industrials')
plt.plot(sp1500_2015['Date'], sp1500_2015['Information_Technology'], label = 'Information Technology')
plt.plot(sp1500_2015['Date'], sp1500_2015['Materials'], label = 'Materials')
plt.plot(sp1500_2015['Date'], sp1500_2015['Real_Estate'], label = 'Real Estate')
plt.plot(sp1500_2015['Date'], sp1500_2015['Utilities'], label = 'Utilities')

plt.legend()
plt.ylabel('Total Return')
plt.xlabel('Date (2015)')
plt.title('Change in Total Return over 2015')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
