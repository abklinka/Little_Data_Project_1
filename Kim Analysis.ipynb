```python
#Re-set columns 
sp1500['Date'] = pd.to_datetime(sp1500['Date'])
sp1500_2016 = sp1500[(sp1500['Date'] >= '6/1/2016') & (sp1500['Date'] <= '11/30/2016')]
sp1500_2016 = sp1500_2016.reset_index(drop=True)


#Rename headers 
sp1500_2016.columns = ['Date', 'Consumer_Discretionary', 'Communication_Services', 'Consumer_Staples', 
                      'Energy', 'Financials', 'Health_Care', 'Industrials', 'Information_Technology',
                      'Materials', 'Real_Estate', 'Utilities']
sp1500_2016.head()

```

```python
plt.plot(sp1500_2016['Date'],sp1500_2016['Consumer_Discretionary'], linewidth = 1)
plt.plot(sp1500_2016['Date'],sp1500_2016['Communication_Services'], linewidth = 1)
plt.plot(sp1500_2016['Date'],sp1500_2016['Consumer_Staples'], linewidth = 1)
plt.plot(sp1500_2016['Date'],sp1500_2016['Energy'], linewidth = 1)
plt.plot(sp1500_2016['Date'],sp1500_2016['Financials'], linewidth = 1)
plt.plot(sp1500_2016['Date'],sp1500_2016['Health_Care'], linewidth = 1)
plt.plot(sp1500_2016['Date'],sp1500_2016['Information_Technology'], linewidth = 1)
plt.plot(sp1500_2016['Date'],sp1500_2016['Industrials'], linewidth = 1)
plt.plot(sp1500_2016['Date'],sp1500_2016['Materials'], linewidth = 1)
plt.plot(sp1500_2016['Date'],sp1500_2016['Real_Estate'], linewidth = 1)
plt.plot(sp1500_2016['Date'],sp1500_2016['Utilities'], linewidth = 1)


plt.title("Hurricane Season 2016")
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.rcParams["figure.figsize"] = (15,5)
plt.show()


```

## Mapping Hurricane Hermine

```python
#calculate 21 days before Active date and 21 days after End date to set as x-axis range

#change format from string to datetime objects 
hurricanes['Active_date'] = pd.to_datetime(hurricanes['Active_date'])
hurricanes['End_date'] = pd.to_datetime(hurricanes['End_date'])

hurricanes2= hurricanes.set_index('Name')
start_date= hurricanes2.loc["Hermine", "Active_date"]
end_date= hurricanes2.loc["Hermine", "End_date"]

#set 3 week delta for x min and max
x_min_herm= start_date - timedelta(days=21)
x_min_herm = x_min_herm.strftime('%m/%d/%Y')

x_max_herm= end_date + timedelta(days=21)
x_max_herm = x_max_herm.strftime('%m/%d/%Y')

#plot for those dates for a chart called Hermine
hurricane_hermine = sp1500_2016[(sp1500_2016['Date']>= x_min_herm)& (sp1500_2016['Date'] <= x_max_herm)]

```

```python
# Plot 11 sectors
fig = plt.figure()
ax = plt.axes()
for x in range(1, 12):
    plt.plot(hurricane_hermine['Date'], hurricane_hermine.iloc[:,x])

    
# Shade area representing Hurricane Hermine 
x_Fmt = DateFormatter("%b-%d")
plt.axvline(start_date, color='teal')
plt.axvline(end_date, color='teal')

ax.xaxis.set_major_formatter(x_Fmt)
ax.axvspan(start_date, end_date, alpha=0.3, color='mediumaquamarine')

plt.ylabel('Total Return')
plt.xlabel('Date')
plt.title('Hurricane Hermine(2016)', fontsize = 20)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.rcParams["figure.figsize"] = (15,5)

#Finesse
sns.set()
plt.show()


```
